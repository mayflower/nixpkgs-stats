```python
import pandas
import plotly.plotly as plot
import cufflinks
cufflinks.go_offline(connected=True)
```

```python
data = pandas.read_csv('./commit_history.csv')
data['date'] = pandas.to_datetime(data['date'])
data = data.loc[~data.message.str.startswith('Merge pull request')]
data
```

```python
data_by_year = data.groupby(data.date.dt.year)
commits_per_year = data_by_year.count().iloc[:,0]
```

```python
commits_per_year.iplot(kind="bar", yTitle="Number of Commits", xTitle="Year", title="Commits/Year")
```

```python
commits_per_year.pct_change().iplot(kind='bar')
```

```python
data_by_month = data.groupby([data.date.dt.year, data.date.dt.month])
```
