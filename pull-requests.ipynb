# Pull Requests

```python
import json
import os
import pandas
from pandas.io.json import json_normalize
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks
init_notebook_mode(connected=True)
pulls = []
for file_name in [f'pulls/{f}' for f in os.listdir('./pulls')]:
    with open(file_name) as f:
        pulls.append(pandas.DataFrame.from_dict(json_normalize(json.load(f))))
raw_data = pandas.concat(pulls).set_index('number')
data = raw_data.loc[:, [
    'additions', 'assignee.login', 'author_association', 'base.ref', 'base.sha', 'body', 'changed_files',
    'closed_at', 'comments', 'commits', 'created_at', 'deletions', 'head.ref', 'head.repo.owner.login',
    'base.sha', 'locked', 'maintainer_can_modify', 'mergeable', 'merge_commit_sha', 'merged_at',
    'merged_by.login', 'rebaseable', 'review_comments', 'state', 'title', 'updated_at', 'user.login'
]]
data.loc[:, 'closed_at'] = pandas.to_datetime(data['closed_at'])
data.loc[:, 'created_at'] = pandas.to_datetime(data['created_at'])
data.loc[:, 'merged_at'] = pandas.to_datetime(data['merged_at'])
data.loc[:, 'updated_at'] = pandas.to_datetime(data['updated_at'])
data.loc[:, 'merge_duration'] = pandas.to_timedelta(data.merged_at - data.created_at)
data
```

## PRs Created per Year

```python
iplot(data.groupby(data.created_at.dt.year).size().iplot(asFigure=True, kind="bar", dimensions=(750, 500)))
```

## Number of PRs by User

```python
iplot(data.groupby('user.login').size().sort_values(ascending=False)[:20].iplot(asFigure=True, kind='bar', dimensions=(750, 500)))
```

## Number of PRs by Assignee

```python
iplot(data.groupby('assignee.login').size().sort_values(ascending=False)[:20].iplot(asFigure=True, kind='bar', dimensions=(750, 500)))
```

## Number of PRs by Merger

```python
iplot(data.groupby('merged_by.login').size().sort_values(ascending=False)[:20].iplot(asFigure=True, kind='bar', dimensions=(750, 500)))
```

## Number of PRs created per Day

```python
daily = data.set_index('created_at').resample('D').size()
monthly_mean = daily.resample('M').mean()
monthly_mean.index = monthly_mean.index.strftime('%Y/%m')
```

```python
iplot(monthly_mean.iplot(asFigure = True, vline=['2017/09', '2017/03', '2016/09', '2016/03', '2015/09', '2014/12', '2014/04', '2013/10'], dimensions=(750, 500)))
```

## Number of PRs opened vs. merged per Day

```python
merged_daily = data.set_index('closed_at').resample('D').size()
merged_monthly_mean = merged_daily.resample('M').mean()
merged_monthly_mean.index = merged_monthly_mean.index.strftime('%Y/%m')
```

```python
iplot((monthly_mean - merged_monthly_mean).iplot(asFigure = True, vline=['2017/09', '2017/03', '2016/09', '2016/03', '2015/09', '2014/12', '2014/04', '2013/10'], dimensions=(750, 500)))
```
